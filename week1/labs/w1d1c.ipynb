## Question 2: Functions!

In programming, a function is a reusable block of code that performs a specific task. <br>
It allows you to group related pieces of code together, give them a name, and call them whenever needed. 

For Example:
``` python
def function_name(parameter1, parameter2, ...):
    """
    Description of the function goes here
    """
    # the code that gets executed when you call the function goes here
    # Parameters are values that can be used in this code 

    # Optionally, output a result using the "return" statement
    
```

Here's a simple example of a function:

````python
def greet_and_get_first_letter(name):
    """
    A function that greets the user and returns the first letter of their name
    """
    print("Hello " + name + "!")
    return name[0]
````
