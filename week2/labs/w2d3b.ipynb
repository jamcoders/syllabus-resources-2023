Suppose we wanted to use this idea of recursion to print a `text` again and again a certain number of times. Let's see how we could do this in the code below.

Read the code below slowly, then run it. **Focus on the line commented: `# THIS LINE CAUSES THE RECURSION`**



````python
def repeat_print(text, how_many_to_print):
    """
    Prints a text a number of times.

    Args:
      text (str): A text to print
      how_many_to_assert_equal (int): Number of times to print the text

    Returns:
      None
    """

    # If we want to print zero times, then just return
    if how_many_to_print == 0:
        return

    # Otherwise...
    else:
        # Print the text once.
        print(text)
        # Since we just printed it out once, we now want to print the text out one less time.
        repeat_print(text, how_many_to_print - 1) # THIS LINE CAUSES THE RECURSION


repeat_print('Hello JamCoder', 3)
repeat_print('Recursion is the best', 8)
````

**Important:** Read the following definitions:
* In recursion, a ***base case*** is an execution of the function in which we do not use the function again.
* A ***recursive case*** is an execution of the function in which we do use the function again.

For example, in the `repeat_print()` function, the base case is when `how_many_to_print == 0`. In that case, we return without calling `repeat_print()` again.

The recursive case is when `how_many_to_print > 0`. In that case, we call `repeat_print(text, how_many_to_print - 1)`
